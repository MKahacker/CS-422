```python
auto = pd.read_table("https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data",sep="\s+", header=None)
auto.columns=["mpg","cylinders", "displacement", "horsepower", "weight", "acceleration","model-year", "origin", "car-name"]
```

```python
auto.describe()
```

```python
auto.head()
```

```python
numerical = auto.drop(columns="car-name")
numerical.describe()
```

```python
numerical["horsepower"] = (numerical["horsepower"].replace(to_replace="?",value=np.nan)).astype(np.float64)
numerical.describe()
```

```python
mean_imp = preprocessing.Imputer(strategy="mean", missing_values=np.nan)
mean = mean_imp.fit_transform(numerical.drop(columns=["mpg","cylinders","displacement",
                                                      "weight","acceleration","model-year","origin"]))
numerical["horsepower"] = mean
numerical.describe()
```

```python
numerical.head()
```

```python
clust = ac(linkage="average", n_clusters=3, compute_full_tree=False)
auto_cluster = clust.fit(numerical)
print(auto_cluster.labels_)
```
